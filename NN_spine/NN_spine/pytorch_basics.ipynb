```python
import time
import numpy as np

size_of_vec = int(1e6)
```

```python
def pure_python_version():
    t1 = time.time()
    X = range(size_of_vec)
    Y = range(size_of_vec)
    Z = [X[i] + Y[i] for i in range(len(X))]
    return time.time() - t1

t1 = pure_python_version()
print(f'Pure python time: {t1}')
```

```python
def numpy_version():
    t1 = time.time()
    X = np.arange(size_of_vec)
    Y = np.arange(size_of_vec)
    Z = X + Y
    return time.time() - t1

t2 = numpy_version()
print(f"numpy version: {t2}")
```
